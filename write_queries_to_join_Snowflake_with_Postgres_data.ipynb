```python
import glaredb
```

```python
con = glaredb.connect()
```

```python
con.sql(
    f"""
    SELECT * FROM
        read_postgres(
        'host=pg.demo.glaredb.com port=5432 user={postgres_user} password={postgres_password} dbname=postgres',  
        'public',
        'borough_lookup'
    ) 
    """
).to_pandas()
```

```python
con.sql(
    f"""
SELECT postgres.borough_name, snowflake.* FROM 
    read_snowflake(
    '{snowflake_account}.us-central1.gcp', 
    '{snowflake_user}',
    '{snowflake_password}',
    'sandbox',
    'compute_wh',
    'accountadmin',
    'public',
    'nyc_sales'
    ) snowflake 
LEFT JOIN 
    read_postgres(
        'host=pg.demo.glaredb.com port=5432 user={postgres_user} password={postgres_password} dbname=postgres',  
        'public',
        'borough_lookup'
    ) postgres ON
    snowflake.borough = postgres.borough_id
    limit 100  
    """).to_pandas()
```
